```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision
from torchvision import transforms
```

```python
import os
train_SEMimglist = os.listdir("/data/jhkim/samsung/AI_challenge_data/Train/SEM")
train_depthimglist = os.listdir("/data/jhkim/samsung/AI_challenge_data/Train/Depth")
len(train_SEMimglist)
len(train_depthimglist)
print(train_SEMimglist[0])
print(train_depthimglist[0])
```

```python
trans = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,),(0.5,))])
trans_wo_Norm = transforms.ToTensor()
```

```python
SEM_root = "/data/jhkim/samsung/AI_challenge_data/Train/SEM/"
depth_root = "/data/jhkim/samsung/AI_challenge_data/Train/Depth/"

class img_dataset(Dataset):
    def __init__(self, file_list, transform):
        self.file_list = file_list
        self.transform = transform
    def __len__(self):
        return len(self.file_list)
    def __getitem__(self, index):
        img_path = self.file_list[index]
        img = Image.open(SEM_root + img_path)
        img_transformed = self.transform(img)

        return img_transformed, img_path
```

```python
img_path = train_SEMimglist[0]
img = Image.open(SEM_root + img_path)
#img_transformed = trans_wo_Norm(img)
img_transformed = np.array(img)
print(img_transformed)
```

```python
img_path = train_SEMimglist[1]
img = Image.open(SEM_root + img_path)
#img_transformed = trans_wo_Norm(img)
img_transformed2 = np.array(img)
print(img_transformed-img_transformed2)
```

```python
img_path = train_depthimglist[0]
img = Image.open(depth_root + img_path)
#img_transformed = trans_wo_Norm(img)
img_transformed = np.array(img)
print(img_transformed[:][20])
```

```python
train_dataset = img_dataset(file_list = train_SEMimglist, transform=trans)
print(len(train_dataset))
train_dataloader = DataLoader(train_dataset, batch_size=3, shuffle=True)

batch_iterator = iter(train_dataloader)
images, img_paths = next(batch_iterator)

print(images.size())
print(len(img_paths))
print(img_path[0])
img_path[0][:-9]+'.png'
```
